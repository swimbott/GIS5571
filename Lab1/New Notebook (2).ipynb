```python
import requests
import pandas
import json

url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?location=44.991192%2C-93.260567&radius=2500&type=restaurant&keyword=subs&key=AIzaSyAOnstvDO861Lpg-4Jw8LlWWEaXEflkQ-4"

# restaurant/sub stations in mlps

payload={}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)

print(response.text)
```

```python
r_json_restaurants=response.json()["results"]
```

```python
df_restaurants=pandas.DataFrame(columns=("Name", "Latitude","Longitude"))

for feature in range(len(r_json_restaurants)):
    name=[r_json_restaurants[feature]["name"]]
    Latitude=[r_json_restaurants[feature]["geometry"]["location"]["lat"]]
    Longitude=[r_json_restaurants[feature]["geometry"]["location"]["lng"]]
    
    df_restaurants.loc[feature]=name+Latitude+Longitude
    
df_restaurants.head()
```
